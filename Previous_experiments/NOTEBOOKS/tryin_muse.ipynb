### get language specific words from test and val

```
get_ipython().system('git clone https://github.com/ftyers/vardial-shared-task.git')

    
def split_words_by_lang(original_filename, languages, target_folder):

    get_ipython().system(f'mkdir {target_folder}')

    lang_files={lang:None for lang in languages}
    for lang in languages:
        fn = f"{target_folder}/{lang}"
        lang_files[lang] = open(fn, 'w', encoding='utf-8')

    with open(original_filename, 'r') as f:
        for line in f:
            lang, wf, _,_,_ = line.split('\t')
            print(wf.strip(), file = lang_files[lang])

    for lang, filename in lang_files.items():
        filename.close()

split_words_by_lang("vardial-shared-task/train/roa-uncovered", langs_to_process["roa"], "train_words")
split_words_by_lang("vardial-shared-task/dev/roa-uncovered", langs_to_process["roa"], "val_uncovered_words")
split_words_by_lang("vardial-shared-task/dev/roa-covered", langs_to_process["roa"], "val_covered_words")
```
